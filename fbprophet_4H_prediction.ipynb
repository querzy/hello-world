```
import pandas as pd
from pandas import read_csv
from pandas import to_datetime
from pandas import DataFrame
from datetime import datetime
from fbprophet import Prophet
from matplotlib import pyplot

def extractDigits(lst): 
    return [[el] for el in lst] 

def get_num(xx):
    return int(''.join(ele for ele in xx if ele.isdigit()))

fileLst=['EURUSD240.csv','GBPUSD240.csv','AUDUSD240.csv','USDJPY240.csv','USDCAD240.csv','NZDUSD240.csv','USDCHF240.csv','GBPJPY240.csv','CADJPY240.csv',
'EURJPY240.csv','EURGBP240.csv','GBPNZD240.csv','AUDJPY240.csv','NZDJPY240.csv','USDTRY240.csv','XAGUSD240.csv','XAUUSD240.csv']


for file in fileLst :
    valperiod=get_num(file)
    filename1 = datetime.now().strftime("%d %b %Y - %H%M")
    filename=file.replace('.csv','')
    parite=filename.replace(str(valperiod),'')
    print(parite)
    print(valperiod)
    print('  ')
    df = pd.read_csv(file)
    #df=df[-500:]
    df.head(5)

    df.describe()

    # Select only the important features i.e. the date and price
    df = df[["Date","Close"]] # select Date and Price# Rename the features: These names are NEEDED for the model fitting
    df = df.rename(columns = {"Date":"ds","Close":"y"}) #renaming the columns of the datasetdata.head(5)

    df.columns =['ds','y']

    df['ds']=to_datetime(df['ds'])

    model=Prophet()

    model.fit(df)

    future = pd.date_range(start ='2021.03.05 10:00',  
            end ='2021.06.05 06:00', freq ='4H') 
      

    future = future.strftime('%Y.%m.%d %H:%M').tolist()

    future = extractDigits(future)

    future=DataFrame(future)

    future.columns=['ds']

    future['ds']=to_datetime(future['ds'])

    forecast = model.predict(future)

    print(forecast[['ds','yhat','yhat_lower','yhat_upper']].head())

    model.plot(forecast)

    pyplot.title(parite+' '+str(valperiod))

    pyplot.show()
```
